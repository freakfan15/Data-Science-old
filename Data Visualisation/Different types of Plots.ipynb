```python
#Line Plot
x = np.arange(10) #creates an array from 0 to 9
y1 = x**2 
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1, color="red",label="Apple",marker="o")
plt.plot(x,y2, color="green", label="Kiwi", marker="*", linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruits over time")
plt.legend() #to show the index in upper left corner
plt.show() #to show the graphs after object creation
```

```python
# to show both graphs in two separate ones
plt.plot(x,y1, color="red")
plt.show()
plt.plot(x,y2, color="green")
plt.show()
```

# Bar Graph

```python
plt.style.use("classic")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25, [10,20,15],width=1/2, label="current year", tick_label=["Gold","Platinum", "Silver"])
plt.bar(x_coordinates+0.25, [20,10,12],width=1/2, label="Next year")
plt.title("Metal Price comparison over two years")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

```python
#same plot in dark background
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25, [10,20,15],width=1/2, label="current year", tick_label=["Gold","Platinum", "Silver"])
plt.bar(x_coordinates+0.25, [20,10,12],width=1/2, label="Next year",color="orange")
plt.title("Metal Price comparison over two years")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.legend()
plt.show()
```

# Pie Chart

```python
subjects = ["Maths", "Chem", "Physics", "English"]
weightage = [20,10,15,5]
# explode splits the part which is marked "1".
# autopct thing shows the percentage of each part
plt.pie(weightage,labels=subjects, explode=(0,0,0.1,0), autopct='%1.1f%%', shadow=True)
plt.title("Subjects")
plt.show()
```

```python
Xsn = np.random.randn(100) #creates 100 random data points from standard normal distribution
sigma =5
u=60
#using np.round to round off the digits into integers
X = np.round(Xsn*sigma +u) # Shifting from S.N to normal. i.e Xsn=(X-u)/sigma

#suppose we want to see marks distribution in physics where mean marks is 70 and deviation is 5 marks
Xph = np.round(Xsn*5 +70)
print(X,Xph) # we can see most of the points lie b/w 55 to 60 [as sigma is 5 & u is 60]
```

# Histogram

```python
plt.style.use("seaborn")
plt.hist(X, label="Maths", alpha=0.9) # .hist for histogram
plt.hist(Xph, label="Physics", alpha=0.8) # alpha is opacity. Adding opacity to view overlapped region
plt.xlabel("Marks Range")
plt.ylabel("Probability / Frequnecy count of Students")
plt.title("Normal distribution of marks in Histogram")
plt.legend()
plt.show()
```

# Movie Data Visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read a CSV file
df = pd.read_csv("movie_metadata.csv")
```

```python
df.head(n=10) #shows first 10 rows
```

```python
df.columns # shows all the columns in the dataset
```

```python
titles = list(df.get('movie_title'))
print(type(titles))
```

```python
print(titles[:5])
```

```python
freq_titles={}

for title in titles:
    length = len(title)
    #if that particular length title is not present in dict: freq_title
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
    #if that length is already present then inc it
        freq_titles[length]+=1
print(freq_titles)        
```

```python
X=np.array(list(freq_titles.keys()))
Y=np.array(list(freq_titles.values()))
print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of movie Title")
plt.ylabel("Freq count")
plt.title("Movie Data Visulaisation")
plt.show()  #plot obtained shows that it is a normal distribution
```
